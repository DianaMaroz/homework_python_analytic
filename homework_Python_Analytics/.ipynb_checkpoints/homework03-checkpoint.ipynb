### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0


```python
df_loc['delta_renovated']=df_loc['yr_renovated']-df_loc['yr_built']
df_loc.sample(10)
```

```python
df_loc.loc[df_loc['yr_renovated'] == 0, 'delta_renovated']= 0
df_loc.sample(10)
```

### 1.3 Создайте признаки года продажи, месяца продажи


```python
def get_year(date):
    return date[:4]
```

```python
df_loc['yr_sail']=df_loc['date'].apply(get_year).astype(int)
df_loc.sample(10)
```

```python
def get_month(date):
    return date[4:6]
```

```python
df_loc['month_sail']=df_loc['date'].apply(get_month).astype(int)
df_loc.sample(10)
```

### 1.4 Удалите признаки date, zipcode, lat, long

```python
df_loc.drop(columns = ['date', 'zipcode', 'lat', 'long'], inplace=True)
df_loc.head()
```

## Задача 3
Составьте несколько сводных таблиц
### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
df_loc.groupby('bedrooms').agg(mean_price = ('price', 'mean')).sort_values('mean_price')
```

### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df_loc.groupby('condition').agg(min_price = ('price', 'min'),
                               mean_price = ('price', 'mean'),
                               max_price = ('price', 'max'))
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
pd.crosstab(index= df_loc['waterfront'], columns = df_loc['view'])
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
pd.crosstab(index= df_loc['bedrooms'], columns = df_loc['floors'])
```

### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома


```python
df_loc.pivot_table(index = df_loc['condition'], 
               columns = df_loc['grade'], 
               values = 'price',
              aggfunc = 'median')
```
